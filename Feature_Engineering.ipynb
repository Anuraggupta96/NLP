```python
#reading dataset
path = 'drive/My Drive/Colab Notebooks/train.csv'
df = pd.read_csv(path)
print("Shape of Original Dataset:",df.shape)
#taking 10% dataset
df=df.sample(frac=0.1)
df=df.reset_index()
df= df.drop(["Unnamed: 0", "index"], axis =1)
print("Shape of sub Dataset:",df.shape)
```

```python
df["Reviews"] = df["Reviews"].convert_dtypes()
df.dtypes
```

```python
df.head()
```

```python
#defining function to remove htmls from data fram
def no_html(x):
    soup=bs(x,'html.parser')
    clean_text= soup.get_text()
    return(clean_text)

#cleaning htmls using efined function
df["Cleaned_text"] = df["Reviews"].apply(lambda y: no_html(y))
```

```python
df.head()
```

```python
#removing newline character
df.Cleaned_text = df.Cleaned_text.str.replace(r'[\"\\...\\\n]', '')
```

## Removing URLs, emails and punctuations

```python
#removing mails
maildf= df
maildf.head()
```

```python
# load model
nlp = spacy.load('en_core_web_sm')
```

```python
def no_email(z):
  doc1 = nlp(z)
  no_email= " ".join([token.text for token in doc1 if not token.like_email])
  return(no_email)
maildf["Cleaned_text"] = maildf["Cleaned_text"].apply(lambda y: no_email(y))
```

```python
#removing urls
urldf= maildf
urldf.head()
```

```python
def no_url(z):
  doc2 = nlp(z)
  no_url= " ".join([token.text for token in doc2 if not token.like_url])
  return(no_url)
urldf["Cleaned_text"] = urldf["Cleaned_text"].apply(lambda y: no_url(y))
```

```python
punc_df= urldf
punc_df.head()
```

```python
#removing punctuations
def no_punc(z):
  doc3 = nlp(z)
  no_punc= " ".join([token.text for token in doc3 if  not token.is_punct])
  return(no_punc)
punc_df["Cleaned_text"] = punc_df["Cleaned_text"].apply(lambda y: no_punc(y))
```

## Feature Engineering

```python
#number of words
finaldf= punc_df
finaldf['word_count'] = finaldf['Cleaned_text'].str.split().str.len()
finaldf.head()
```

```python
#number of character with space
finaldf['num_chr'] = finaldf['Cleaned_text'].str.len()
finaldf.head()
```

```python
#number of character without space
finaldf['Cleaned_text1'] = finaldf['Cleaned_text'].str.replace(' ','')
finaldf['num_chr_wo_space'] = finaldf['Cleaned_text1'].str.len()
finaldf= finaldf.drop("Cleaned_text1", axis=1)
finaldf.head()
```

```python
#avg word length
finaldf['avg_len']= finaldf['num_chr_wo_space']/finaldf['word_count']
finaldf.head()
```

```python
#count of numbers
sampledf= finaldf
#removing punctuations
def num_token(z):
  doc4 = nlp(z)
  num_token1= [token.text for token in doc4 if (token.like_num and token.is_digit)]
  return(num_token1)
sampledf["Cleaned_num"] = sampledf["Cleaned_text"].apply(lambda y: num_token(y))
```

```python
sampledf["num_count"]= sampledf['Cleaned_num'].str.len()
finaldf= sampledf.drop("Cleaned_num", axis=1)
```
